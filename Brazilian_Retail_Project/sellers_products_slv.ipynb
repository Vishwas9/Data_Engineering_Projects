```python
sellers_df = spark.read.table("projects.project_1.bronze_sellers")
products_df = spark.read.table("projects.project_1.bronze_products")
```

```python
sellers_df = sellers_df.drop("ingested_at")
sellers_df.write.format("delta")\
    .mode("overWrite")\
    .saveAsTable("projects.project_1.slv_sellers")
```

```python
products_df.drop("ingested_at")
products_df.write.format("delta")\
    .mode("overWrite")\
    .saveAsTable("projects.project_1.slv_products")
```
