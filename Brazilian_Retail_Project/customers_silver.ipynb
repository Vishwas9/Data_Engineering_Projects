```python
from pyspark.sql.functions import *
customer_df = spark.read.table("projects.project_1.bronze_customers")
display(customer_df.count())
```

```python
customer_df= customer_df.dropDuplicates()
customer_df = customer_df.withColumn("customer_city", initcap("customer_city"))
customer_df_new = customer_df.drop("ingested_at")
display(customer_df_new.count())
```

```python
customer_df_new.write.format("delta")\
    .mode("overwrite")\
    .option("overwriteSchema", "true")\
    .saveAsTable("projects.project_1.slv_customers")
```
