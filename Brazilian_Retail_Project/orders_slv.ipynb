```python
from pyspark.sql.functions import *
orders_df = spark.read.table("projects.project_1.bronze_orders")
order_items_df = spark.read.table("projects.project_1.bronze_order_items")
order_review_df = spark.read.table("projects.project_1.order_reviews_bronze")
order_payment_df = spark.read.table("projects.project_1.bronze_order_payments")
```

```python
orders_df.fillna("", subset = ['order_delivered_carrier_date', 'order_delivered_customer_date'])
orders_df.dropDuplicates().count()
orders_df.drop("ingested_at")
```

```python
orders_new_df = orders_df.drop("ingested_at").withColumns({
    "order_purchase_timestamp" : expr("try_cast(order_purchase_timestamp AS date)"),
    "order_approved_at" : expr("try_cast(order_approved_at AS date)"),
    "order_delivered_carrier_date" : expr("try_cast(order_delivered_carrier_date AS date)"),
    "order_delivered_customer_date" : expr("try_cast(order_delivered_customer_date AS date)"),
    "order_estimated_delivery_date" : expr("try_cast(order_estimated_delivery_date AS date)")
})
display(orders_new_df)
```

```python
order_items_df = order_items_df.withColumn("shipping_limit_date", expr("try_cast(shipping_limit_date AS date)")).drop("ingested_at")
display(order_items_df)
```
