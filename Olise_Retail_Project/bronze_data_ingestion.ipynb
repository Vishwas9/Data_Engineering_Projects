```python
from pyspark.sql.functions import *
```

## Customer Data

```python
# Ingesting Data From Volumes
raw_customers_df = spark.read.format("csv")\
        .option("header", "true")\
        .option("InferSchema", "true")\
        .load("/Volumes/projects/project_1/landing/olist_customers_dataset.csv")
```

```python
#writing the customers data into a delta table
bronze_customers_df = raw_customers_df.withColumn("ingested_at", current_timestamp())
bronze_customers_df.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("projects.project_1.bronze_customers")
```
